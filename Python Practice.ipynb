#PATTERN MATCHING WITH REGULAR EXPRESSIONS
##Finding patterns of text without regular expressions




```python
#we are creating a function to check whether a strong matches the pattern of a phone number


def isPhoneNumber(text):
  if len(text) != 12:
    return False
  for i in range(0,3):
    if not text[i].isdecimal():
      return False
  if text[3] != '-':
    return False
  for i in range(4,7):
    if not text[i].isdecimal():
      return False
  if text[7] != '-':
    return False
  for i in range(8,12):
    if not text[i].isdecimal():
      return False
  return True

  print(isPhoneNumber(text))
```

```python
text = input("enter a phone number")
print(isPhoneNumber(text))
```

```python
isPhoneNumber("222-111-3333")
```

```python
print('415-555-4242 is a phone number:')
print(isPhoneNumber('415-555-4242'))
print('Moshi moshi is a phone number:')
print(isPhoneNumber('Moshi moshi'))
```

```python
#Build a function to fetch phone number from a string.

def fetch_number(text):
  for i in range(len(text)):
    chunk = text[i:i+12]
    if isPhoneNumber(chunk):
      print("Phone number found:", chunk)
      print("Done")

```

```python
fetch_number("My phone number is 123-234-2345")
```

```python
#lets import regex and create regex object

import regex as re
```

```python
phonenumRegex = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')
```

Regex search function helps serching string and find a match else returns false
Regex match objects have a group function which returns the matched text from the string.

```python
mo = phonenumRegex.search('My number is 813-067-0022 and 111-222-3333')
print("Number found: "+ mo.group())

```

```python
#Grouping with Parentheses

phonenumRegex = re.compile(r'(\d\d\d)-(\d\d\d-\d\d\d\d)')


mo = phonenumRegex.search("My number is 423-234-2323.")


mo.group(1)

```

```python
mo.group(2)
```

```python
mo.group(0)
```

```python
mo.group()
```

```python
mo.groups()
```

```python
areacode, mainnumber = mo.groups()

print(areacode, mainnumber)
```

```python
#escape character
phonenumRegex = re.compile(r'(\(\d\d\d\))-(\d\d\d-\d\d\d\d)')


mo = phonenumRegex.search("My number is (423)-234-2323.")


mo.group(1,2,0)

```

```python
#Matching multiple groups with pipe
heroregex = re.compile(r'Batman|Superman')

mo1 = heroregex.search('Batman is not as good as Superman.')

mo1.group()
```

```python
mo2 = heroregex.search('Superman is an alien not human.')
mo2.group()
```

```python
batRegex = re.compile(r'Bat(man|toy|car|copter|mobile)')

mo = batRegex.search('Batmobile lost a wheel')
mo.group()
```

```python
mo.group(1)
```

```python
 batRegex = re.compile(r'Bat(wo)?man')

 mo1 = batRegex.search('Batwoman and Batman are true friends.')
 mo1.group()

 #(wo)? part of the regular expression means that the pattern wo is an optional group. The regex will match text that has zero instances or one instance of wo in it. This is why the regex matches both 'Batwoman' and 'Batman'
```

```python
phoneRegex = re.compile(r'(\d\d\d-)?\d\d\d-\d\d\d\d')
mo1 = phoneRegex.search('My number is 415-555-4242')
mo1.group()
```

```python
mo2 = phoneRegex.search('My number is 555-4242')
mo2.group()
```
